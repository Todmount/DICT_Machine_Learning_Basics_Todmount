```python
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
# Data
features, labels = make_blobs(n_samples = 1000,
                              n_features = 2,
                              centers = 3,
                              cluster_std =2.5,
                              shuffle = True,
                              random_state = 16)

# Visualize data
fig, ax = plt.subplots()
ax.scatter(features[:, 0], features[:, 1], c=labels)
plt.show()

# Model
model = KMeans(n_clusters=3, random_state=1).fit(features)

# Labels by clusters
predicted = model.labels_

# Evaluate
score = silhouette_score(features, predicted)
print('Score is ', score)

# Visualize predictions
fig, ax = plt.subplots()
ax.scatter(features[:, 0], features[:, 1], c=predicted)
plt.show()
```
