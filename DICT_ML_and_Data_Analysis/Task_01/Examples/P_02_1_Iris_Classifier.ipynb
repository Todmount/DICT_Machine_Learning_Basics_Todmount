```python
# Data: the Iris dataset: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#examples-using-sklearn-datasets-load-iris
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
```

```python
samples = [5, 55, 125]

for i in samples:
  print(f'{i}-th sample is "{iris.target_names[iris.target[i]]}" <--> {iris.data[i]}')
```

```python
iris.feature_names
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])
_ = ax.legend(
    scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes"
)
```

```python
# Dummy classifier: https://scikit-learn.org/stable/modules/generated/sklearn.dummy.DummyClassifier.html
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import train_test_split

# x, y
features, target = iris.data, iris.target

# Split data
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=0)

# Classifier
dummy = DummyClassifier(strategy='uniform', random_state=1)

# Train/fit classifier
dummy.fit(features_train, target_train)

print(f'Dummy classifier score: {dummy.score(features_test, target_test)}')
```

```python
# Other classifier - Random forest: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html#sklearn-ensemble-randomforestclassifier
from sklearn.ensemble import RandomForestClassifier

# Classifier
classifier = RandomForestClassifier()

# Train/fit classifier
classifier.fit(features_train, target_train)

print(f'Random Forest Classifier classifier score: {classifier.score(features_test, target_test)}')
```
