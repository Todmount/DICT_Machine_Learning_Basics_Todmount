## Провести аналіз класифікаторів DummyClassifier, Logistic Regression та Random Forest Classifier на прикладі набору даних Digits

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
class_names = digits.target_names
```

### Dummy

```python
dummy = DummyClassifier(strategy='stratified')
# dummy = DummyClassifier(strategy='uniform') # ~ Accuracy: 0.1056
dummy.fit(X_train, y_train)
dummy_pred = dummy.predict(X_test)
```

```python
# Evaluate and visualize results
accuracy_dummy = accuracy_score(y_test, dummy_pred)
confusion_mat_dummy = confusion_matrix(y_test, dummy_pred)
classification_rep_dummy = classification_report(y_test, dummy_pred)

print("\nDummy Classifier Results:")
print(f"Accuracy: {accuracy_dummy:.4f}")
print("\nClassification Report:")
print(classification_rep_dummy)
```

```python
# Plot confusion matrix
sns.heatmap(confusion_mat_dummy, annot=True, fmt="d", cmap="Blues", cbar=None, xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.title("Confusion Matrix - Dummy Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

### Logistic Regression

```python
logistic_reg = LogisticRegression(max_iter=1000)
logistic_reg.fit(X_train, y_train)
logistic_pred = logistic_reg.predict(X_test)
```

```python
# Evaluate and visualize results
accuracy_logistic = accuracy_score(y_test, logistic_pred)
confusion_mat_logistic = confusion_matrix(y_test, logistic_pred)
classification_rep_logistic = classification_report(y_test, logistic_pred)

print("\nLogistic Regression Results:")
print(f"Accuracy: {accuracy_logistic:.4f}")
print("\nClassification Report:")
print(classification_rep_logistic)
```

```python
# Plot confusion matrix
sns.heatmap(confusion_mat_logistic, annot=True, fmt="d", cmap="Blues", cbar=None, xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```

### Random Forest Classifier

```python
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest.fit(X_train, y_train)
rf_pred = random_forest.predict(X_test)
```

```python
# Evaluate and visualize results
accuracy_rf = accuracy_score(y_test, rf_pred)
confusion_mat_rf = confusion_matrix(y_test, rf_pred)
classification_rep_rf = classification_report(y_test, rf_pred)

print("\nRandom Forest Classifier Results:")
print(f"Accuracy: {accuracy_rf:.4f}")
print("\nClassification Report:")
print(classification_rep_rf)
```

```python
# Plot confusion matrix
sns.heatmap(confusion_mat_rf, annot=True, fmt="d", cmap="Blues", cbar=None, xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.title("Confusion Matrix - Random Forest Classifier")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
