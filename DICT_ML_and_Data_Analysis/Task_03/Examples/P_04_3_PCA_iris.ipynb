```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
data = datasets.load_iris()

features = StandardScaler().fit_transform(data.data)

labels = data.target

names = data.target_names
```

```python
pca = PCA(n_components=3)

features_pca = pca.fit_transform(features)

print('Explained variances: ', pca.explained_variance_)

print('Explained variances (ratio): ', pca.explained_variance_ratio_)
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(features_pca[:, 0], features_pca[:, 2], c=labels)
_ = ax.legend(
    scatter.legend_elements()[0], names, loc="upper right", title="Classes"
)
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

ax.scatter(
    features_pca[:, 0],
    features_pca[:, 1],
    features_pca[:, 2],
    c=labels,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

plt.show()
```
