```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
data = datasets.load_digits(n_class=10)

features = StandardScaler().fit_transform(data.data)

labels = data.target

names = data.target_names

print(data.data.shape)
```

```python
# PCA

pca = PCA(n_components=2)

features_pca = pca.fit_transform(features)

print('Explained variances: ', pca.explained_variance_)

print('Explained variances (ratio): ', pca.explained_variance_ratio_.sum())

_, ax = plt.subplots()
scatter = ax.scatter(features_pca[:, 0], features_pca[:, 1], c=labels)
_ = ax.legend(
    scatter.legend_elements()[0], names, loc="upper right", title="Classes"
)
```

```python
# tSNE

from sklearn.manifold import TSNE

tsne = TSNE(n_components=2,
        n_iter=500,
        n_iter_without_progress=250,
        n_jobs=2,
        random_state=10)

features_tsne = tsne.fit_transform(features)

_, ax = plt.subplots()
scatter = ax.scatter(features_tsne[:, 0], features_tsne[:, 1], c=labels)
_ = ax.legend(
    scatter.legend_elements()[0], names, loc="upper right", title="Classes"
)
```
