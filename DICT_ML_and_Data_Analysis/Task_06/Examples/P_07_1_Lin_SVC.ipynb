```python
# Imports
from sklearn.svm import LinearSVC # Linear kernel SVM
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data
iris = datasets.load_iris()
```

```python
# Labels
iris.target
```

```python
# Find two classes subset
limit = np.where(iris.target == 2)[0][0]

scaler = StandardScaler()

features = scaler.fit_transform(iris.data[:limit, :2])
labels = iris.target[:limit]
names = iris.target_names[:limit]

_, ax = plt.subplots()
scatter = ax.scatter(features[:, 0], features[:, 1], c=labels)
ax.set(xlabel=names[0], ylabel=names[1])
_ = ax.legend(scatter.legend_elements()[0], names, loc="lower right", title="Classes")
```

```python
# support vector classifier (SVC)
svc = LinearSVC(C=1.0)

# fit model
model = svc.fit(features, labels)
```

```python
# Visualize

# Data
_, ax = plt.subplots()
scatter = ax.scatter(features[:, 0], features[:, 1], c=labels)
ax.set(xlabel=names[0], ylabel=names[1])
_ = ax.legend(scatter.legend_elements()[0], names, loc="lower right", title="Classes")

# Line
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0]) / w[1]

# Add line
plt.plot(xx, yy)
plt.show()
```

```python
# Data
observations = [[ -2, 3], [0, 0]]

classes = svc.predict(observations)

print('Classes:', classes)

for i in range(len(observations)):
  print(observations[i], f'is {names[classes[i]]}')
```
